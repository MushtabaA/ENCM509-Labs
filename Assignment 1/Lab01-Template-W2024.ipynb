Before plotting the function $f(x)$, we will first plot the original function $sin(x)$ using the interval $x\ \epsilon \ (0,2\pi)$:

```python
x = np.linspace(0, 2*np.pi)
y = np.sin(x)
```

Using the interval: $x\ \epsilon \ (-6,6)$, we will now plot the function $f(x)$ below using three different combinations of markers, line styles, and colours.
$$f(x) = sin x + \frac{cos(4x)}{4} + \frac{sin(5x)}{5} + \frac{cos(6x)}{6}$$

```python
x = np.linspace(-6, 6)
y = np.sin(x) + (np.cos(4*x)/4) + (np.sin(5*x)/5) + (np.cos(6*x)/6)
```

##### Combination #1 (Marker "o", Line Style "--", and Colour "red"):

```python
plt.plot(x, y, marker='o', linestyle='--', color='red')
plt.title('Function f(x)')
plt.xlabel('X')
plt.ylabel('$f(X)$')
```

##### Combination #2 (Marker "^", Line Style ":", and Colour "gray"):

```python
plt.plot(x, y, marker='^', linestyle=':', color='gray')
plt.title('Function f(x)')
plt.xlabel('X')
plt.ylabel('$f(X)$')
```

##### Combination #3 (Marker "1", Line Style "-.", and Colour "magenta"):

```python
plt.plot(x, y, marker='1', linestyle='-.', color='magenta')
plt.title('Function f(x)')
plt.xlabel('X')
plt.ylabel('$f(X)$')
```

#### Analysis Of Data:

As shown in the plots above, we were able to display the function $f(x)$ in a variety of ways through different markers, line styles, and colours. In addition, we can also see the hints of similarities between function $f(x)$ and the original $sin(x)$ plot since the function $f(x)$ includes the original $sin(x)$ plot. Since the function $f(x)$ is essentially a mix between $sin(x)$ and some other $cos(x)$ functions, we get a "weird" but "nice looking" output with sudden "drops" and "spikes" that we can plot.

## **Exercise 4**:

Using the age demograhic data from the previous exercise, we can also plot a normal distribution chart using the formula below with the help of the library "scipy":
$$f(x) = \frac{1}{\sigma \sqrt{2\pi}}e^{-\frac{1}{2}(\frac{x-\mu}{\sigma})^{2}}$$
where $\sigma$ is the standard deviation and $\mu$ is the mean.

```python
# counting of people at each age interval shown in age labels
ages = [16, 23, 347, 428, 334, 398, 923, 1015, 216, 11]
age_labels = ['0-9','10-19','20-29','30-39','40-49','50-59','60-69','70-79','80-89','90-99']
```

First, we will create a new vector corresponding to each age interval which we will then use to calculate both the mean and standard deviation.

```python
vec = []

index = 0
for age in ages:
    for count in range(age):
        vec.append(index)

    index += 1
```

Next, we will utilize np.mean and np.std provided by numpy to calcuate both the mean and standard deviation.

```python
mean = np.mean(vec)
stddev = np.std(vec)
print("Mean:", mean, "Standard Deviation:", stddev)
```

Using numpy, we found that the mean was $\mu = 5.33$, while the standard deviation was $\sigma = 1.84$.
\
Now that we have the mean and standard deviation, we can plot the normal distribution using scipy's "norm" probability density function below:

```python
# Smoothing out the vector (evenly spaced values within a given interval)
vec = np.arange(min(vec), max(vec), 0.01)
plt.plot(vec, norm.pdf(vec, mean, stddev))
plt.xticks(np.arange(len(age_labels)), age_labels, rotation=45)
plt.xlabel('Age', fontweight='bold')
plt.ylabel('Probability Density', fontweight='bold')
plt.title('Normal Distribution of Age on Cruise Ship', fontweight='bold')
```

#### Analysis Of Data:

After adding the age interval labels to the normal distribution plot above, it is clearly evident that a large majority of the data falls in the age interval of 40 to 79 years old. This means that statistically speaking, there is a high probability of an individual on the Diamond Princess Cruise ship falling within that age range due to the area under the curve within that interval. In addition, this age interval also has the highest probability density as shown by the y-axis in the plot, indicating where the age demographic data is most concentrated, which in this case, is at its highest at the age intervals of "50-59" and "60-69".
