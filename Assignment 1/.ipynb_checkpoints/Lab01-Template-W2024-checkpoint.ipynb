# Plotting with Matplotlib

```python
import numpy as np
import matplotlib.pyplot as plt

x = np.arange(0, 2, 0.01)
plt.plot(x,x)
```

```python
x = np.linspace(0, 2*np.pi)
y = np.sin(x)
```

```python
plt.plot(x,y, '-ro')
plt.title('sin function')
plt.xlabel('X')
plt.ylabel('$sin(X)$')
```
