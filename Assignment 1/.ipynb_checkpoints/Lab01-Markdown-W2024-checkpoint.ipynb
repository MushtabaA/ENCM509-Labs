## Lists

Unordered (bulleted):
* first item
* second item
* third item


Ordered:
1. first item
2. second item
3. third item

Nested lists:
* first item:
    * sub-first item
    * sub-first item again
* second item:
    1. sub-second item
    * sub-second item (without number!)
    * sub-second item (you just need to put the first number)
