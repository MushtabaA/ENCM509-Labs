First, we will plot a histogram and normal distribution curve using a set of 80 samples and 10 bins with $\mu = 50$ and $\sigma = 5$

```python
# Theorectical Mean and Stddev
mu = 50
std = 5

# Drawing 80 samples from a normal distribution
pop_norm = norm.rvs(mu, std, size=80)
pop_norm
```

```python
# Theorectical Mean and Stddev
mu = 50
std = 5

# Drawing 1000 samples from a normal distribution
pop_norm = norm.rvs(mu, std, size=1000)
pop_norm
```

```python
# Plotting a histogram considering 100 bins of data
plt.hist(pop_norm, bins=100, density=True);
plt.title('Histogram - 100 Bins of Data')
print('REAL mean: %.2f; std: %.2f' % (np.mean(pop_norm), np.std(pop_norm)))
```

In order to get a better side-by-side visual, we will also plot both of the histograms together

```python
mu = 50
std = 5

norm1 = norm.rvs(mu, std, size=80)
norm2 = norm.rvs(mu, std, size=1000)

plt.figure(figsize=(12,4))
plt.subplot(1,2,1)
plt.hist(norm1, bins=10, density=True);
plt.hist(norm2, bins=100, density=True, color='orange', alpha=0.8)
plt.title('Histograms')
plt.legend(['Samples: 80, Bins: 10', 'Samples: 1000, Bins: 100'])
```

#### Analysis Of Data:

Although we kept the mean $(\mu)$ and standard deviation $(\sigma)$ the same for both plots, it is clear that an increase in sample size and the number of bins causes a large difference in the distribution of the data. For instance, it is evident that by increasing the sample size, we are able to now view a much larger distribution/dispersion of the data including outliers on both sides of the mean even though the standard deviation remained the same (as shown in the orange overlapping histogram above). In addition, by increasing the bin size, we are also increasing the number of intervals/groups that the data can fall in which is necessary especially when increasing the sample size. This is because if the number of bins were too small or too large, the distribution/dispersion of the data would not be easily seen, and thus, the distribution trend in the data would be difficult to perceive.

## **Exercise 2:**

First, we will load the data for an "imposter" writing of the word `Biometrics`

```python
# data folder:
# './SampleData/biometrics/'
# './SampleData/calgary/'
# './SampleData/hello/'

# 'genuine' or 'impostor'
subdir = 'impostor/'
data_dir = './SampleData/biometrics/' + subdir

csv_file = pd.read_csv(data_dir + '1.csv')
```

After loading the data, we can visualize the first lines of data loaded which contains 4 columns [X, Y, Time, Pressure] which we will also split into separate variables

```python
# splitting the original data/matrix into 3 variables
coord1 = csv_file[['X', 'Y']].to_numpy()
time1 = csv_file['Time'].to_numpy().reshape((-1,1))
prs1 = csv_file['Pressure'].to_numpy().reshape((-1,1))

csv_file.head()
```

Now we will set up the matplotlib colormap in order to plot the word `Biometrics` in color representative of pressure

```python
# to correctly map the pressure into the colormap the normalization is required 
prs1 = prs1 / np.max(prs1) * 255
prs1 = prs1.astype(int)

pressuremap = cm.get_cmap('jet', np.max(prs1)+1)
```

After setting up the colormap, we can now plot how the word `Biometrics` was written in 2D with the pressure applied being represented by color `(Red = high pressure, Blue = low pressure)`

```python
for i in range(len(prs1)-1):
    c = pressuremap(prs1[i]).squeeze()
    
    time_diff = abs(time1[i+1] - time1[i])
    if time_diff == 0:
        time_diff = 0.0001
        
    if time_diff < 17: # You may need to change this threshould
        im = plt.plot(coord1[i:i+2,0], -coord1[i:i+2,1], c=c)

plt.xlabel('X')
plt.ylabel('Y')
plt.title('Pen pressure')

sm = cm.ScalarMappable(cmap='jet', norm=plt.Normalize(vmin=0, vmax=1))
plt.colorbar(sm);
```

Before plotting the histogram and normal distribution, we will first calculate the mean $(\mu)$ and standard deviation $(\sigma)$ of the pressure for the word `Biometrics`

```python
# mean and standard deviation of the presssure
mu_prs = np.mean(prs1)
std_prs = np.std(prs1)
print("Mean:", mu_prs, "Standard Deviation:", std_prs)
```

Now that we have both the mean $(\mu)$ and standard deviation $(\sigma)$, we can plot both the histogram and normal distribution

```python
# Plotting the histogram of pressure
plt.figure(figsize=(14,6))
plt.subplot(1,2,1)
plt.hist(prs1, density=True);
plt.xlabel('Pressure', fontweight='bold')
plt.ylabel('Probability', fontweight='bold')
plt.title('Histogram', fontweight='bold')

# Plotting the normal distribution of presssure using scipy's probability density function
plt.subplot(1,2,2)
x = np.arange(0, 300, 0.01)
x_pdf = norm.pdf(x, loc=mu_prs, scale=std_prs)
plt.xlabel('Pressure', fontweight='bold')
plt.ylabel('Probability Density', fontweight='bold')
plt.title('Normal Distribution', fontweight='bold')
plt.plot(x, x_pdf);
```

#### Analysis Of Data:

As shown in both the histogram and the normal distribution plots above, it is evident that most of the data is heavily distributed/dispersed in the high pressure range (150 - 250). This makes sense as we calculated the mean of the written word `Biometrics` to be $\mu = 210.89$. Moreover, as exhibited previously in the 2D colour plot, most of the letters (if not all) were written with high pressure (with the exception of small details such as the dot on top of the letter "i" and the tail-ends of the letters "t", "c", and "s"), which further explains the reasoning behind the normal distribution curved being skewed right to the high pressure range.

## **Exercise 4:**

In order to perform statistical analysis of pressure and velocity distribution, we'll use average values of pressure and velocity to calculate the mean $(\mu)$ and standard deviation $(\sigma)$ of both in order to plot normal distributions

We'll use the function below to calculate the velocity for each signature of the genuine word `Labs` by using the time and coordinates fields

```python
def calc_velocity(time, coords):
    vel = np.zeros((len(time)-1, 1))

    for i in range(len(time)-1):
        distance = np.sqrt((coords[i+1,0] - coords[i,0])**2 +
                           (coords[i+1,1] - coords[i,1])**2)
        
        # if too fast time_diff = 0
        time_diff = time[i+1] - time[i]
        if time_diff == 0:
            time_diff = 0.0001

        vel[i] = distance / time_diff
        vel[i] = int(vel[i] * 1000) + 1

    vel = np.insert(vel, 0, 1)
    vel = vel / np.max(vel) * 255
    vel = vel.astype(int)
    
    return vel
```

Next we will store the mean $(\mu)$ and standard deviation $(\sigma)$ of both the pressure and velocity after loading up all the samples of the word `Labs`

```python
# columns: mean_vel, std_vec, mean_prs, std_prs
stats_measures = np.zeros((30,4))
data_dir = './Data/Usman-Labs/'

# using 30 signatures
for i in range(30):
    csv_file = pd.read_csv(data_dir + '%d.csv' % (i+1))
    
    coord1 = csv_file[['X', 'Y']].to_numpy()
    time1 = csv_file['Time'].to_numpy().reshape((-1,1))
    prs1 = csv_file['Pressure'].to_numpy().reshape((-1,1))

    vel = calc_velocity(time1, coord1)
    
    stats_measures[i, 0] = np.mean(vel)
    stats_measures[i, 1] = np.std(vel)
    
    stats_measures[i, 2] = np.mean(prs1)
    stats_measures[i, 3] = np.std(prs1)
    
    print('#%d - pressure: %.2f; velocity: %.2f' %\
          (i, np.mean(prs1), np.mean(vel)))
```

Now that we have both the mean $(\mu)$ and standard deviation $(\sigma)$ of both the pressure and velocity, we can now plot normal distributions of the average pressure and velocity values across the 30 samples of the word `Labs`

```python
# Plotting the histogram of velocity
plt.figure(figsize=(14,6))
plt.subplot(1,2,1)
plt.hist(stats_measures[:, 0], density=True);
plt.xlabel('Velocity', fontweight='bold')
plt.ylabel('Probability', fontweight='bold')
plt.title('Velocity Histogram', fontweight='bold')

# Plotting the normal distribution of velocity using scipy's probability density function
plt.subplot(1,2,2)
x = np.arange(min(stats_measures[:, 0]), max(stats_measures[:, 0]), 0.01)
x_pdf = norm.pdf(x, loc=np.mean(stats_measures[:, 0]), scale=np.std(stats_measures[:, 0]))
plt.xlabel('Velocity', fontweight='bold')
plt.ylabel('Probability Density', fontweight='bold')
plt.title('Normal Distribution of Average Velocity Values', fontweight='bold')
plt.plot(x, x_pdf);
```

```python
# Plotting the histogram of pressure
plt.figure(figsize=(14,6))
plt.subplot(1,2,1)
plt.hist(stats_measures[:, 2], density=True);
plt.xlabel('Pressure', fontweight='bold')
plt.ylabel('Probability', fontweight='bold')
plt.title('Pressure Histogram', fontweight='bold')

# Plotting the normal distribution of pressure using scipy's probability density function
plt.subplot(1,2,2)
x = np.arange(min(stats_measures[:, 2]), max(stats_measures[:, 2]), 0.00125)
x_pdf = norm.pdf(x, loc=np.mean(stats_measures[:, 2]), scale=np.std(stats_measures[:, 2]))
plt.xlabel('Pressure', fontweight='bold')
plt.ylabel('Probability Density', fontweight='bold')
plt.title('Normal Distribution of Average Pressure Values', fontweight='bold')
plt.plot(x, x_pdf);
```

#### Analysis Of Data:

After plotting both the histograms and normal distributions of both velocity and pressure, we were able to see some interesting trends/characteristics for the genuine signature of the word `Labs`. For instance, the mean of the velocity was $\mu = 20.86$, which translates to the normal distribution of the velocity, as most of the velocity values had the highest probability density in that area. Furthermore, the mean of the pressure was $\mu = 0.66$, and as shown in the normal distribution of the pressure above, most of the pressure values fall in this area, which means that the writer of the genuine word `Labs` applies medium-high pressure when writing.
